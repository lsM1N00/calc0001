```
# 문장 분리를 한 것
from pytube import YouTube
from youtube_transcript_api import YouTubeTranscriptApi
from kiwipiepy import Kiwi

def youtube_info(youtube_url):
    try:
        yt = YouTube(youtube_url)
        video_id = yt.video_id
        return video_id
    except Exception as e:
        return None, str(e)
# api 사용
def get_script_from_api(video_id):
    transcription = YouTubeTranscriptApi.get_transcript(video_id, languages=['ko'])

    segments_with_durations = []
    text = ''

    for segment in transcription:
        text += segment['text'] + ' '

    text_ = text.replace('\n', ' ')

    kiwi = Kiwi()
    separate = kiwi.split_into_sents(text_)

    # 자막과 문장 수 비교
    if len(separate) != len(transcription):
        print(f"Warning: Number of sentences ({len(separate)}) does not match number of transcriptions ({len(transcription)}).")

    # 최대 반복 범위 결정
    iteration_range = min(len(separate), len(transcription))

    for i in range(iteration_range):
        start_time = transcription[i]['start']
        end_time = start_time + transcription[i]['duration']

        # 시간 계산
        start_min = int(start_time // 60 )
        start_sec = int(start_time % 60 )
        end_min = int(end_time // 60 )
        end_sec = int(end_time % 60 )

        segments_with_durations.append({
            'text': separate[i].text if i < len(separate) else "텍스트 없음",
            'start_time': (start_min, start_sec),
            'end_time': (end_min, end_sec)
        })

        # 시간과 텍스트 출력
        print(f"{start_min:02d}:{start_sec:02d} ~ {end_min:02d}:{end_sec:02d}: {separate[i].text if i < len(separate) else '텍스트 없음'}")

    return segments_with_durations

# 유튜브 링크 입력
youtube_url = input("유튜브 링크를 입력: ")
video_id = youtube_info(youtube_url)

if video_id:
    print(f"비디오 ID: {video_id}")
    print("스크립트 응답:")
    api_response = get_script_from_api(video_id)
else:
    print("에러 발생: 비디오 ID를 추출할 수 없습니다.")
```
